```python
from numpy import exp, log, log2, log10
import matplotlib.pyplot as plt
```

```python
MAX_IN_LINE = 200
MAX_FLOW_PER_SECOND = 50

x = [i for i in range(1, MAX_IN_LINE + 1)]
```

```python
plt.figure(figsize = (15, 7))
plt.grid()

plt.plot(x, [log2((i/MAX_IN_LINE) + 1)*MAX_FLOW_PER_SECOND for i in range(1, MAX_IN_LINE+1)], label = "normal")

plt.plot(x, [log2((i/MAX_IN_LINE)**(1/2) + 1)*MAX_FLOW_PER_SECOND for i in range(1, MAX_IN_LINE+1)], label = "^0.5")
plt.plot(x, [log2((i/MAX_IN_LINE)**(1/3) + 1)*MAX_FLOW_PER_SECOND for i in range(1, MAX_IN_LINE+1)], label = "^0.3")
plt.plot(x, [log2((i/MAX_IN_LINE)**(1/4) + 1)*MAX_FLOW_PER_SECOND for i in range(1, MAX_IN_LINE+1)], label = "^0.25")
plt.plot(x, [log2((i/MAX_IN_LINE)**(1/5) + 1)*MAX_FLOW_PER_SECOND for i in range(1, MAX_IN_LINE+1)], label = "^0.2")
plt.plot(x, [log2((i/MAX_IN_LINE)**(1/6) + 1)*MAX_FLOW_PER_SECOND for i in range(1, MAX_IN_LINE+1)], label = "^0.16")

plt.plot(x, [log2((i/MAX_IN_LINE)**2 + 1)*MAX_FLOW_PER_SECOND for i in range(1, MAX_IN_LINE+1)], label = "^2")
plt.plot(x, [log2((i/MAX_IN_LINE)**3 + 1)*MAX_FLOW_PER_SECOND for i in range(1, MAX_IN_LINE+1)], label = "^3")
plt.plot(x, [log2((i/MAX_IN_LINE)**4 + 1)*MAX_FLOW_PER_SECOND for i in range(1, MAX_IN_LINE+1)], label = "^4")
plt.plot(x, [log2((i/MAX_IN_LINE)**5 + 1)*MAX_FLOW_PER_SECOND for i in range(1, MAX_IN_LINE+1)], label = "^5")
plt.plot(x, [log2((i/MAX_IN_LINE)**6 + 1)*MAX_FLOW_PER_SECOND for i in range(1, MAX_IN_LINE+1)], label = "^6")

plt.legend()
```

### To stay longer < 1

```python
plt.figure(figsize = (15, 7))
plt.grid()

plt.plot(x, [log2((i/MAX_IN_LINE)**10 + 1)*MAX_FLOW_PER_SECOND for i in range(1, MAX_IN_LINE+1)], label = "^10")
plt.plot(x, [log2((i/MAX_IN_LINE)**15 + 1)*MAX_FLOW_PER_SECOND for i in range(1, MAX_IN_LINE+1)], label = "^15")
plt.plot(x, [log2((i/MAX_IN_LINE)**20 + 1)*MAX_FLOW_PER_SECOND for i in range(1, MAX_IN_LINE+1)], label = "^20")
plt.plot(x, [log2((i/MAX_IN_LINE)**25 + 1)*MAX_FLOW_PER_SECOND for i in range(1, MAX_IN_LINE+1)], label = "^25")



plt.legend()
```
