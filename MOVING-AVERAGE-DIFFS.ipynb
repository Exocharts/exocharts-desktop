```python
import pandas as pd
import numpy as np
from finta import TA
```

```python
period = 15
```

```python
ethpath = "Binance_ETHUSDT_2022_minute.csv"
eth = pd.read_csv(ethpath, skiprows = 1, parse_dates=["date"])[::-1].set_index("date")
```

```python
df = pd.DataFrame.from_dict({
    "ema" : TA.EMA(eth, period),
    "tema" : TA.TEMA(eth, period),
    "price" : (eth.open + eth.close ) / 2,
    "sma" : TA.SMA(eth, period = period),
})
```

### First 300 rows

```python
df.head(300).plot(figsize = (15, 5))
```

### Last 300 rows

```python
df.tail(300).plot(figsize = (15, 5))
```

```python
print("EMA error: ", np.mean(np.abs(df.ema - df.price)))
```

```python
print("TEMA error: ", np.mean(np.abs(df.tema - df.price)))
```

```python
print("SMA error: ", np.mean(np.abs(df.sma - df.price)))
```
